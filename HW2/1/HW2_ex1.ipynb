```python
import numpy as np
import matplotlib.pyplot as plt


def f_1(x1, x2):

    return 4*np.power(x1,2) - np.multiply(x1,x2) + 7 - x1 - x2

def dx1(x1, x2):

    return 8*x1 - x2 - 1

def dx2(x1, x2):

    return -1*x1 - 1

x1 = np.arange(-0.5, 1, 0.001)
x2 = np.arange(-0.5, 1, 0.001)

X1, X2 = np.meshgrid(x1, x2)
Z = f_1(X1, X2)

lr = 0.01

x1_init = 0
x2_init = 0

X1_trajectory = []
X2_trajectory = []

grad_mag_x1 = []
grad_mag_x2 = []

X1_trajectory.append(x1_init)
X2_trajectory.append(x2_init)

for i in range(4):
    
    grad_x1 = lr*dx1(x1_init, x2_init)
    grad_x2 = lr*dx2(x1_init, x2_init)
    x1_new = x1_init -grad_x1
    x2_new = x2_init -grad_x2

    x1_init = x1_new
    x2_init = x2_new

    grad_mag_x1.append(grad_x1)
    grad_mag_x2.append(grad_x2)

    if i == 3:
        grad_x1 = lr*dx1(x1_init, x2_init)
        grad_x2 = lr*dx2(x1_init, x2_init)
        grad_mag_x1.append(grad_x1)
        grad_mag_x2.append(grad_x2)
    
    X1_trajectory.append(x1_init)
    X2_trajectory.append(x2_init)



fig = plt.figure(figsize = (16,8))
ax = fig.add_subplot(1, 2, 1)
ax.contour(X1, X2, Z)
ax.quiver(X1_trajectory, X2_trajectory, grad_mag_x1, grad_mag_x2, scale_units = 'xy', angles = 'xy', scale = 1, color = 'r', alpha = .9)
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_surface(X1, X2, Z,rstride = 5, cstride = 5, cmap = 'jet', alpha=0.5)
ax.plot3D(X1_trajectory, X2_trajectory, f_1(X1_trajectory, X2_trajectory), marker = '*', color = 'r', alpha = .4, label = 'Gradient descent')
plt.show()
```

```python
import copy
import numpy as np

class tanh:

    def __init__(self):
        pass

    def forward(self, f_in):
        self.output = np.tanh(f_in)
    
    def backward(self, b_in):
        return np.multiply(1 - np.power(self.output, 2), b_in)
    
class sigmoid:

    def __init__(self):
        pass

    def forward(self, f_in):
        self.output = 1/(1 + np.exp(-f_in))
    
    def backward(self, b_in):
        return np.multiply(b_in, np.multiply(self.output, 1-self.output))


class BCE_loss:

    def __init__(self):
        pass

    def forward(self, y_true, y_predicted):
        y_predicted = np.clip(y_predicted, 0.000001, 0.999999)
        self.output = np.mean(-(y_true*np.log(y_predicted)+(1-y_true)*np.log(1-y_predicted)))

    def backward(self,output, label):

        return -(label/output)+((1-label)/(1-output))
    




inputs = np.random.uniform(-2, 1, (2, 10))
outputs = np.random.randint(0, 2, (1, 10))

W1 = np.random.uniform(-2,2,(3,2))
W2 = np.random.uniform(-2,2,(1,3))
b1 = np.random.uniform(-2,2,(3,1))
b2 = np.random.uniform(-2,2,(1,1))

activation1 = tanh()
activation2 = sigmoid()

lr = 0.1
Loss = BCE_loss()

for epoch in range(10001):

    # SGD optimizer
    for i in range(inputs.shape[1]):

        x = inputs[:,i]
        d = outputs[:,i]

        # forward propagation
        n1 = np.matmul(W1, x) + b1[:,0]
        activation1.forward(n1)
        n2 = np.matmul(W2, activation1.output) + b2[:,0]
        activation2.forward(n2)
        y_predict = copy.copy(activation2.output)
        Loss.forward(d, activation2.output)

        # Backward Propagation

        e = d - activation2.output
        loss_derivative1 = Loss.backward(activation2.output, d)
        s2 = activation2.backward(loss_derivative1)

        loss_derivative2 = np.matmul(W2.T, s2)
        s1 = activation1.backward(loss_derivative2)

        # updating weights and biases

        W1 = W1 - lr*np.matmul(s1.reshape((3,1)),x.reshape((1,2)))
        W2 = W2 - lr*s2*activation1.output.T
        b1 = b1 - lr*s1
        b2 = b2 - lr*s2
    
    
    y_predict[np.where(y_predict>0.5)] = 1
    y_predict[np.where(y_predict<=0.5)] = 0
    accuracy = np.mean(outputs == y_predict)
    if epoch % 100 == 0:
        print(f'Epoch:{epoch}')
        print(f'loss:{Loss.output}')
        print(f'Accuracy: {np.mean(accuracy)}')
        if accuracy == 1:
            break
        
        print('--------------------------')
    

```

```python
import copy
import numpy as np

class Linear:

    def __init__(self):
        pass

    def forward(self, f_in):
        self.output = f_in
    
    def backward(self, b_in):
        return b_in
    
class ReLU:

    def __init__(self):
        pass

    def forward(self, f_in):
        self.output = np.maximum(0,f_in)
        self.input = f_in
    
    def backward(self, b_in):
        self.b_output = b_in
        self.b_output[self.input<=0] = 0
        return self.b_output


class MSE_loss:

    def __init__(self):
        pass

    def forward(self, y_true, y_predicted):
        self.output = np.mean((y_true-y_predicted)**2,axis = 0)

    def backward(self,output, label):

        return -2*(label-output)

P = np.array([[-1, 0,  2, -1],
              [-4, 2, -3, 1]])

outputs = np.array([[2.2, 5.7, -2.9, 1.8]])

W1 = np.random.normal(0,1,(3,2))
W2 = np.random.normal(0,1,(1,3))
W_bypass = np.random.normal(0,1,(1,1))
b1 = np.random.normal(0,1,(3,1))
b2 = np.random.normal(0,1,(1,1))

activation1 = ReLU()
activation2 = Linear()

lr = 0.01
Loss = MSE_loss()

for epoch in range(100):

    count = 0
    # SGD optimizer
    for i in range(P.shape[1]):

        x = P[:,i]
        d = outputs[:,i]

        n1 = np.matmul(W1, x) + b1[:,0]
        skip = W_bypass*x[0]
        activation1.forward(n1)
        n2 = np.matmul(W2, activation1.output) + b2[:,0]
        activation2.forward(n2+skip)
        if np.abs(activation2.output - outputs[0,i]) <= 0.1:
            
            count += 1 
        Loss.forward(d, activation2.output)

        
        # Backward Propagation

        e = d - activation2.output
        loss_derivative1 = Loss.backward(activation2.output, d)
        s2 = activation2.backward(loss_derivative1)


        loss_derivative2 = np.matmul(W2.T, s2)
        s1 = activation1.backward(loss_derivative2)
        

        # updating weights and biases
        
        
        W1 = W1 - lr*np.matmul(s1.reshape((3,1)),x.reshape(1,2))
        W2 = W2 - lr*s2*activation1.output.T
        W_bypass = W_bypass - lr*s2*x[0]
        b1 = b1 - lr*s1
        b2 = b2 - lr*s2

    accuracy = count/P.shape[0]

    
    print(f"Loss = {Loss.output}")
    print(f"accuracy = {count/P.shape[1]}")
    print(f"epoch: {epoch}")

    print("--------------------------------------------")
    
    
```
